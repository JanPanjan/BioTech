```python
from Bio import Align
import Bio.SeqIO as SeqIO
aligner = Align.PairwiseAligner()
```

```python
aligner.mode = 'global'
print(aligner)
```

```python
s1 = "AB982279.1.fasta"
s2 = "NW_025235126.1[4419..5271].fasta"
```

```python
with open(s1, 'r') as fs1, open(s2, 'r') as fs2:
    cs1 = SeqIO.parse(fs1, "fasta")
    cs1 = next(cs1)
    cs2 = SeqIO.parse(fs2, "fasta")
    cs2 = next(cs2)
```

```python
print(cs1)
print("seq len:", len(cs1.seq))
```

```python
print(cs2)
print("seq len:", len(cs2.seq))
```

```python
alignment = aligner.align(cs2, cs1)
# scores = aligner.score(cs2, cs1)
```

```python
# print(scores)
```

```python
a = alignment[0]
print("score:", a.score)
print("length:", a.length)
print(a)
```
